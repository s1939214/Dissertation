```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plots
import matplotlib.dates as mdates
from datetime import date as datetime
import pandas as pd

Data = np.loadtxt('Iceland_IMO_C1_95-onwards.txt')
Data_filtered = Data[np.logical_and(Data[:,1]>=63.54,Data[:,1]<=64.45),:]
Data_filtered1= Data_filtered[np.logical_and(Data_filtered[:,2]<=-15.08,Data_filtered[:,2]>=-18.06),:]
Data_filtered2 = Data_filtered1[np.logical_and(Data_filtered1[:,4]>=2.3,Data_filtered1[:,4]<=4),:]
Pre2004 = Data_filtered2[np.logical_and(Data_filtered2[:,0]>=729766,Data_filtered2[:,0]<=731884),:]
post2004 = Data_filtered2[np.logical_and(Data_filtered2[:,0]>=731891,Data_filtered2[:,0]<=734285),:]
post2011 = Data_filtered2[np.logical_and(Data_filtered2[:,0]>=734285,Data_filtered2[:,0]<=737821),:]
total = Data_filtered2[np.logical_and(Data_filtered2[:,0]>=729766,Data_filtered2[:,0]<=737821),:]

Mpre2004 = np.cumsum(10**(2/3*Pre2004[:,4]+16.05))
Mpost2004 = np.cumsum(10**(2/3*post2004[:,4]+16.05))
Mpost2011 = np.cumsum(10**(2/3*post2011[:,4]+16.05))
Dpre2004 = Pre2004[:,0]
Dpost2004 = post2004[:,0]
Dpost2011 = post2011[:,0]
Mtotal = np.cumsum(10**(2/3*total[:,4]+16.05))
Dtotal = total[:,0]

fig, ax = plots.subplots(figsize=(10,10))

plots.plot(Dpre2004-729766, Mpre2004, label='Pre-2004 Eruption: 2118 days')
plots.plot(Dpost2004-731891, Mpost2004, label="Post-2004 Eruption: 2394 days ")
plots.plot(Dpost2011-734285, Mpost2011, label='Post-2011 Eruption 3536')
ax.yaxis.tick_right()
plots.suptitle('Vatnajokull: Cumulative Seismic Moment (M>2.3 & M<3.5)',fontsize=20, weight='extra bold', y=0.96)
ax.set_title('Eruption Periods: 18/12/1998-12/01/1999, 01/11/2004-07/11/2004, 21/05/2011-28/05/2011', fontsize=19, pad=17)
plots.legend(fontsize=15, loc='upper left')
plots.xlabel('Days since last eruption', fontsize=15)
plots.ylabel('Cumul. seis. mo ($\mathregular{Nm 10^{19}})$', fontsize=15);
plots.savefig('Cumulative seismic moment (pdf)')
plots.text(1900,3e+19,"07/11/2004")
plots.text(2120,6.2e+19,"28/05/2011")
plots.text(3150,2.7e+19,"31/01/2021")
plots.text(-100, 5.18e+19, 'Region: $\mathregular{64.27-64.54^{o}}N$; $\mathregular{-17 - -17.5^{o}}W$', fontsize=14)
plots.rcParams.update({'font.size': 14})
plots.grid()
```
