```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.patches as mpatches
Data = pd.read_csv('MassLossTimeSeries.csv')
Data2010 = pd.read_csv('MassLossTimeSeries2010.csv')
```

```python
plot.subplots(2,1,figsize=(20,30))

plot.subplot(211)
CryoDates = Data.CryoSat_Dates
Cryomassloss = Data.CryoSat_massloss
Glaciologytime = Data.Glaciology_time
GlaciologySMB = Data.Glaciology_SMB
Gtime = Data.GRACE_time
GML = Data.Glaciology_massloss
Glaciologytime1 = Data.Glaciology_time1
GSMBinternal = Data.Glaciology_SMBinternal
plot.plot(Glaciologytime, GlaciologySMB, label='Modelled SMB')
plot.plot(CryoDates, Cryomassloss, label='CryoSat')
plot.plot(Gtime, GML, label='GRACE')
plot.plot(Glaciologytime1, GSMBinternal, label='Modelled SMB + Basal melt');
plot.legend(fontsize=25);
plot.grid()
plot.xlabel('Year', fontsize=30)
plot.ylabel('Mass Loss (m)', fontsize=30)
plot.title('Graph showing time series of Mass Loss on Vatnajokull', fontsize=40)
plot.axvline(2011.5, color='r', linestyle='dashed')
plot.text(2011.8, -30, 'Grimsvotn Eruption 2011', fontsize=25)
plot.text(1988, 35, 'A', fontsize=45)
plot.rcParams['font.size'] = 20
rect=mpatches.Rectangle((2009.5,-65),12,-45, 
                        fill=False,
                        color="purple",
                       linewidth=5)
                       #facecolor="red")
plot.gca().add_patch(rect)

plot.subplot(212)
CryoDates2010 = Data2010.CryoSat_Dates
Cryomassloss2010 = Data2010.CryoSat_massloss
Glaciologytime2010 = Data2010.Glaciology_time
GlaciologySMB2010 = Data2010.Glaciology_SMB
Gtime2010 = Data2010.GRACE_time
GML2010 = Data2010.Glaciology_massloss
Glaciologytime12010 = Data2010.Glaciology_time1
GSMBinternal2010 = Data2010.Glaciology_SMBinternal
plot.plot(Glaciologytime2010, GlaciologySMB2010, label='Modelled SMB')
plot.plot(CryoDates2010, Cryomassloss2010, label='CryoSat')
plot.plot(Gtime2010, GML2010, label='GRACE')
plot.plot(Glaciologytime12010, GSMBinternal2010, label='Modelled SMB + Basal melt');
plot.legend(fontsize=25);
plot.grid()
plot.text(2007.6, -52, 'B', fontsize=45)
plot.xlabel('Year', fontsize=30)
plot.ylabel('Mass Loss (m)', fontsize=30)
plot.title('Graph showing zoomed in time series of Mass Loss on Vatnajokull from 2010', fontsize=40)
rect=mpatches.Rectangle((2009.5,-65),12,-45, 
                        fill=False,
                        color="purple",
                       linewidth=5)
                       #facecolor="red")
plot.gca().add_patch(rect)
plot.axvline(2011.5, color='r', linestyle='dashed')
plot.text(2011.8, -50, 'Grimsvotn Eruption 2011', fontsize=25)

plot.savefig('Massloss')
```
