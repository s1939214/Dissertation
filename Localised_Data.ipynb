```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

dataframe = {'year': ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019'],
             'magnitude_sum': ['1.5080213254183657e+19', '0', '5.979524734045527e+17', '1.9311609462850506e+18', '2.0169023718595192e+18', '2.7838043963637494e+18', '5.537749841508885e+17', '2.8985758026539617e+18', '1.5127301137184635e+18']}
df = pd.DataFrame(dataframe)
df['magnitude_sum'] = pd.to_numeric(df['magnitude_sum'])

fig, ax = plt.subplots(figsize=(10,10))
ax.bar(df['year'], df['magnitude_sum'], color='gray', edgecolor='blue')

import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array(df.index).reshape((-1, 1))
y = df['magnitude_sum']
model = LinearRegression().fit(x, y)
r2 = model.score(x, y)

ax.plot(df['year'], model.predict(x), color='red', label=f'R2={r2:.2f}')

line2 = Line2D([0], [0], label='R2 = -0.20', color='red')

ax.legend(handles=[line2], fontsize=13)


ax.set_xlabel('Year', fontsize=15)
ax.set_ylabel('seis. mo ($\mathregular{Nm 10^{19}})$', fontsize=15)
ax.set_title('Grimsvotn Seismic Moment Sum by Year', fontsize=18, weight = 'extra bold')

plt.rcParams.update({'font.size': 14})
plt.text(-0.8, 1.54e+19, 'Region - Grimsvotn basin: $\mathregular{64.27-64.54^{o}}N$; $\mathregular{-17 - -17.5^{o}}W$', fontsize=14)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)


```

```python
import numpy as np
import matplotlib.pyplot as plt

year = ['2011', '2012', '2013', '2014', '2015', '2016']
basal_melt = [55, 44, 31, 33, 28, 26]
magnitude_sum = [1.5080213254183657e+19, 0, 5.979524734045527e+17, 1.9311609462850506e+18, 2.0169023718595192e+18, 2.7838043963637494e+18]

bar_width = 0.35
fig, ax1 = plt.subplots(figsize=(10,10))

# plot the magnitude sum data on the primary y-axis
magnitude_sum_bars = ax1.bar(np.arange(len(year)), magnitude_sum, bar_width, label='Seismic moment Sum', color='red')
ax1.set_ylabel('Seismic Moment Sum')
ax1.ticklabel_format(style='sci', axis='y', scilimits=(0,0)) # format y-axis in scientific notation
ax1.set_xlabel('Year')
ax1.set_xticks(np.arange(len(year)))
ax1.set_xticklabels(year)
ax1.tick_params(axis='y', labelcolor='r') # set y-axis label color to blue
# create a twin axes object with the same x-axis and plot basal melt data on it
ax2 = ax1.twinx()
basal_melt_bars = ax2.bar(np.arange(len(year))+bar_width, basal_melt, bar_width, label='Modelled Basal Melt', color='blue')
ax2.set_ylabel('Basal Melt ($\mathregular{(m) w.e.a^{-1}}$)')
ax2.tick_params(axis='y', labelcolor='blue') # set y-axis label color to blue

# add legend for both sets of bars
bars = [magnitude_sum_bars, basal_melt_bars]
labels = [bar.get_label() for bar in bars]
ax1.legend(bars, labels, bbox_to_anchor=(1.05, 1.0), loc='upper left')

plt.title('Modelled Basal Melt compared to Seismic moment Sum per Year', fontsize=16)
plt.show()

```
