```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plots
import matplotlib.dates as mdates
import datetime as date
Data = np.loadtxt('Iceland_IMO_C1_95-onwards.txt')
Data_filtered = Data[np.logical_and(Data[:,1]>64.27,Data[:,1]<64.54),:]
Data_filtered1= Data_filtered[np.logical_and(Data_filtered[:,2]<-17,Data_filtered[:,2]>-17.5),:]
```

```python
Data_filtered2 = Data_filtered1[np.logical_and(Data_filtered1[:,4]>-1,Data_filtered1[:,4]<10),:]
old_ordinal = Data_filtered2[:,0] 
new_ordinal = old_ordinal + mdates.date2num(np.datetime64('0000-12-31'))
dates = plt.dates.num2date(new_ordinal)
```
